# Task
Generate a bar chart showing the average of each column in the dataframe `df`, excluding the 'actividad' column.

## Calculate column averages

### Subtask:
Calculate the average of each column in the dataframe `df`, excluding the 'actividad' column.


**Reasoning**:
Calculate the mean of each numerical column in the DataFrame, excluding the 'actividad' column, and store the results.



```python
column_averages = df.drop(columns=['actividad']).mean()
print(column_averages)
```

## Prepare data for plotting

### Subtask:
Organize the calculated averages into a format suitable for plotting.


**Reasoning**:
Convert the pandas Series of column averages into a DataFrame and format it for plotting by renaming columns.



```python
df_averages = column_averages.reset_index()
df_averages.columns = ['Course', 'Average Score']
display(df_averages)
```

## Generate bar chart

### Subtask:
Create a bar chart to visualize the average of each column.


**Reasoning**:
Generate a bar chart using the prepared DataFrame to visualize the average scores per course.



```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.bar(df_averages['Course'], df_averages['Average Score'])
plt.xlabel('Course')
plt.ylabel('Average Score')
plt.title('Average Scores per Course')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## Summary:

### Data Analysis Key Findings
